## Step 2: Implement a simple ranking system
One way of recommending movies is to recommend movies that are generally agreed upon to be good. But how do we measure the "goodness" or "likability" of a movie?

Suppose that we simple count the number of likes for each movie and take the 25 movies with the most number of likes. Would this be a good idea?

To help you extract the movie information from the two databases, we have built you a helper function.

```python
#--------  movie_stats
# A function that extracts the number of likes and total number of ratings for a movie
# Input: 
#      movie_name (an optional parameter containing the exact name of the movie)
#      movie_name_contains (an optional parameter containing a portion of the name of the movie)
# Returns: 
#      total_ratings (the total number of ratings for a movie)
#      likes (the total number of likes for a movie)

def movie_stats(movie_name=None, movie_name_contains=None):
    
    #If given an exact movie name:
    if movie_name is not None:        
        #Find the index of the movie, by name, in the "names" dataframe
        movie_index = names_df[names_df['movie_name'] == movie_name].index[0]
        #Get the id for the movie in the "names" dataframe
        movie_id = names_df.loc[movie_index, 'movie_id']
        #Get all ratings for the movie, by id, in the "ratings" dataframe
        ratings_for_movie = ratings_df[ratings_df['movie_id'] == movie_id]
        #Count the total number of ratings
        total_ratings = len(ratings_for_movie)
        #Count the likes (the 1's)
        likes = ratings_for_movie['rating'].sum()
    
    #Otherwise, if given a partial movie name:
    elif movie_name_contains is not None:
        #Find the index of the movie, by name, in the "names" dataframe
        movie_index = names_df[names_df['movie_name'].str.contains(movie_name_contains)].index[0]
        #Get the id for the movie in the "names" dataframe
        movie_id = names_df.loc[movie_index, 'movie_id']
        #Get all ratings for the movie, by id, in the "ratings" dataframe
        ratings_for_movie = ratings_df[ratings_df['movie_id'] == movie_id]
        #Count the total number of ratings
        total_ratings = len(ratings_for_movie)
        #Count the likes (the 1's)
        likes = ratings_for_movie['rating'].sum()
    
    else:
        total_ratings = 0.
        likes = 0.
    
    return float(total_ratings), likes
```

In the following, we've provide you a couple of functions for visualize beta priors and approximating their associated posteriors.

```python
#--------  plot_beta_prior
# A function to visualize a beta pdf on a set of axes
# Input: 
#      a (parameter controlling shape of beta prior)
#      b (parameter controlling shape of beta prior)
#      color (color of beta pdf)
#      ax (axes on which to plot pdf)
# Returns: 
#      ax (axes with plot of beta pdf)

def plot_beta_prior(a, b, color, ax):
    rv = sp.stats.beta(a, b)
    x = np.linspace(0, 1, 100)
    ax.plot(x, rv.pdf(x), '-', lw=2, color=color, label='a=' + str(a) + ', b=' + str(b))
    ax.set_title('Beta prior with a=' + str(a) + ', b=' + str(b))
    ax.legend(loc='best')
    return ax
```

```python
#--------  find_mode
# A function that approximates the mode of a distribution given a sample from the distribution
# Input: 
#      values (samples from the distribution)
#      num_bins (number of bins to use in approximating histogram)
# Returns: 
#      mode (the approximate mode of the distribution)

def find_mode(values, num_bins):
    
    #Make an approximation (histogram) of the distribution using the samples
    bins, edges = np.histogram(values, bins=num_bins)
    #Find the bin in the histogram with the max height
    max_height_index = np.argmax(bins)
    #Find the sample corresponding to the bin with the max height (the mode)
    mode = (edges[max_height_index] + edges[max_height_index + 1]) / 2.
    
    return mode
```

```python
#A list of beta distribution shapes to try out
beta_shapes = [(1, 1), (0.5, 0.5), (2, 2), (4, 2), (2, 4)]
#Length of the list of shapes
n = len(beta_shapes)

#Plot all the beta pdfs in a row
fig, ax = plt.subplots(1, n, figsize=(20, 4))

#Start the index of the current subplot at 0
ax_ind = 0
#Iterate through all the shapes
for a, b in beta_shapes:
    #Plot the beta pdf for a particular shape
    plot_beta_prior(a, b, 'blue', ax[ax_ind])
    #Increment the subplot index
    ax_ind += 1
    
plt.tight_layout()    
plt.show() 
```

**How might adding a prior to our model benifit us in our specific task? Why are beta distributions appropriate priors for our application?**

Beta priors encode commonsensical values about how the likelihood affects the posterior. It encodes a belief about the likability of a movie instead of assuming that all movies are equally likeable.

**What kind of plain-English prior beliefs about the movie does each beta pdf encode?**

The beta priors can be interpreted as follows:
- a = b = 1 suggests that there is not prior knowledge about the movie
- a = b = 0.5, suggests that a movie is believed to be highly polarising (either liked or disliked but nowhere in the middle)
- a = b = 2 suggests that most viewers are likely to be ambivalent about a movie
- a = 4, b = 2 suggests the movie is likely to be liked
- a = 2, b = 4 suggests the movie is likely to be disliked


We will now explore the effect of the priors on the posterior (i.e. the likelihood of a particular "likelihood" given a set of observed ratings).

```python
#--------  sample_posterior
# A function that samples points from the posterior over a movie's 
# likability, given a binomial likelihood function and beta prior
# Input: 
#      a (parameter controlling shape of beta prior)
#      b (parameter controlling shape of beta prior)
#      likes (the number of likes in likelihood)
#      ratings (total number of ratings in likelihood)
#      n_samples (number of samples to take from posterior)
# Returns: 
#      post_samples (a array of points from the posterior)

def sample_posterior(a, b, likes, ratings, n_samples):
    post_samples = np.random.beta(a + likes, b + ratings - likes, n_samples)
    return post_samples
```

The effect of the priors on the posterior can be explored - this effectively gives the likelihood of a particular "likelihood" given a set of observed ratings.

The posterior distribution of a binomial likelihood and a beta prior is also a beta distribution in which the alpha parameter is increased by the number of likes and the beta parameter is increased by the number of dislikes. 

```python
def plot_posterior(movie_index):
    #Get the name of the movie in the top 25 list with index movie_index
    movie_name = top_25_movies[movie_index][0]

    #Get the ratings info for the movie with index movie_index in the top 25 list
    likes = top_25_movies[movie_index][1]
    total_ratings = top_25_movies[movie_index][2]
    likability = top_25_movies[movie_index][3]

    #Print movie info
    print '{}: {} ({}/{})'.format(movie_name, likability, likes, total_ratings)

    #Number of samples to use when approximating our posterior
    n_samples = 10000

    #Plot the posterior corresponding to each prior
    fig, ax = plt.subplots(1, n, figsize=(20, 4))

    #Start the index of the current subplot at 0
    ax_ind = 0

    #Iterate through all the shapes
    for a, b in beta_shapes:   
        #Draw samples from the posterior corresponding to a particular beta prior
        post_samples = sample_posterior(a, b, likes, total_ratings, n_samples)
        #Approximate the posterior with a histogram of these samples
        ax[ax_ind].hist(post_samples, bins=30, color='red', alpha=0.5)
        #Find the approximate mode of the posterior
        mode = find_mode(post_samples, 30)
        #Plot the mode as a vertical line
        ax[ax_ind].axvline(x=mode, linewidth=3, label='Posterior mode')

        #Set title, legends etc
        ax[ax_ind].set_title('Posterior, with Beta prior (a={}, b={})'.format(a, b))
        ax[ax_ind].legend(loc='best')
        #Increment the subplot index
        ax_ind += 1

    plt.tight_layout()
    plt.show() 
```

```python
plot_posterior(0)
plot_posterior(4)
```

**How do we interpret the posterior mode? What does the mode tell us about the likability of a particular movie?
Do the posterior mode tell us the same thing about a movie's likability as the like-percentage? Which metric is more realistic or reasonable?**

The mode of a the posterior distribution is given by:

$$ \frac{\alpha + N_{likes} -1}{\alpha + N_{likes} + \beta + N_{dislikes} - 2}$$

where $N_{likes}$ and $N_{dislikes}$ are the number of likes and dislikes respectively. This is simiar to the like percentage but taking into account the prior belief. 

**How does the choice of prior affect the posterior distribution of the 'likability' for the movies: *Toy Story, Star Wars, The Shawshank Redemption, Down Periscope and Chain Reaction**

The effect of the priors on the posterior can be interpreted for the movie French Twist:
1. With no prior belief the posterior is just equal to the maximum likelihood estimate which suggests that the movie is highly likeable and so the mode of the posterior is 1
2. With the polarising prior, as there are a large percentage of likes, the posterior suggests that the movie is lovable
3. With an ambivalent prior, the probability of the movie is between 0.5 and 1 - i.e. the prior suggested a viewer would be ambivalent but the ratings showed high percentage of likes so the probability is increased from the prior beief.
4. With a positively skewed prior, the ratings reinforce the belief and so the posterior probability of liking a movie is also high. 
5. With a negatively skewed initial prior, the ratings actually show that many people liked the movie, so the probability increases closer to 0.5.  

**How does the effect of the prior on the posterior distribution vary with the number of user ratings?**

The prior adds a bias into calculation of the probability of likeability. If the total number of observations is small, the prior has a bigger effect on the resulting posterior distribution. If the total number of observations is large, there is a higher confidence in the result and the prior has a minor effect i.e. if a movie has not been rated by many people, the information encoded in the prior belief is required much more than if the movie has already been verified by a large number of viewers. From examining the mode of the beta distribution it can be seen that when the number of observations is large, the mode is similar to the like percentage. 

A symmetric prior (one that suggest that a movie is equally likely to be rated highly vs badly) is an appropriate choice given no previous knowledge about the movie or a user. This biases the posterior distribution to the center, so  requires more positive observations to shift the mode to the left or right. The mode therefore takes into account the both the percentage of positive reviews as well as the total number of reviews, which gives a better predction of movie likeability compared to simply using like percentage.

Three predictive models can be fit to the data to see which has the best predictive accuracy over the years:

1. Logistic Regression with three binary classifiers (one vs rest)
2. LDA
3. Adjusted LDA taking into account the estimated yearly changes in proportions of the demographic groups. 

The model can be trained on the the 2000 dataset and then used to predict on all the subsequent years. An adjusted LDA model can be built by incorporating the knowledge about the estimated yearly changes in proportions of the three demographic groups into the prior for the model  - the prior can then be updated for each testing set. 

The adjusted LDA model can be calculated by considering the Bayes classifier. The Bayes classifier assigns an observation X = x to the class $k$ which has the largest value of:

$$ P(Y=k|X=x) = \frac{\pi_kf_k(x)}{\sum_{l=1}^{K}\pi_lf_l(x)}$$

where $fk(X) ≡ Pr(X = x|Y = k)$ is the the density function of X for an observation that comes from the kth class and is asssumed to tbe Gaussian. The log probability is given by:

$$δk(x) = x^T Σ^{-1}μ_k − \frac{1}{2}μ_k^T Σ^{-1}μ_k + log (π_k)$$

which is known as the linear discriminant function. This posterior probability can be calculated from the LDA function in sklearn using predict_proba or predict_log_proba and then incorporating the prior probability based on the information about changing demographics i.e. the proportions either increase by 25% for class 2 or decrease by 12.5% 

```python
data_2000 = np.loadtxt('datasets/dataset_1_year_2000.txt')
x_2000 = data_2000[:, 0:-1]
y_2000 = data_2000[:,-1]

## Logistic Model

log_model = LogReg()
log_model.fit(x_2000, y_2000)

# LDA model

lda_model = LDA()
lda_model.fit(x_2000, y_2000)

# Initialise empty arrays to save the scores

logreg_scores = np.zeros((10, 1))
lda_scores = np.zeros((10, 1))
lda_adjusted_scores = np.zeros((10, 1))

for i in range(1,11):
    data_year = np.loadtxt('datasets/dataset_1_year_' + str(2000+i) + '.txt')
    x_year = data_year[:,0:-1]
    y_year = data_year[:,-1]
    logreg_scores[i-1] = log_model.score(x_year, y_year)
    lda_scores[i-1] = lda_model.score(x_year, y_year)
    
    # calculate predicted probabilities and multiply by the prior 
    # -> could also use log probs and add the log of the prior
    # need to multiply the change (either 0.875 or 1.25) by the number of years from the base year
    
    lda_probs = lda_model.predict_proba(x_year)
    lda_probs[:,0] = lda_probs[:,0]*(0.875**(i-1)) # the class 0 prior probability is 1*0.875^n (n years from 2000)
    lda_probs[:,1] = lda_probs[:,1]*(0.875**(i-1))
    lda_probs[:,2] = lda_probs[:,2]*(1.25**(i-1)) # the class 2 prior probability is 1*1.25^n (n years from 2000)
    
    y_predict = np.argmax(lda_probs, axis = 1)    
    lda_adjusted_scores[i-1] = np.sum(y_predict == y_year)/float(len(y_year))

years = np.arange(2001,2011)
w = 0.25
plt.figure(figsize = (30,10))
plt.bar(years, logreg_scores, w, color = 'r', label = 'Logistic Regresssion')
plt.bar(years + w, lda_scores, w, color = 'g', label = 'LDA')
plt.bar(years+ 2*w, lda_adjusted_scores, w, color = 'b', label = 'LDA adjusted')

plt.xlabel('Year', fontsize=30)
plt.ylabel('Accuracy Score', fontsize=30)
plt.legend(loc='best', bbox_to_anchor=(1,0.5), fontsize=30)
plt.yticks(fontsize=30)
plt.xticks(years + 1.5*w , years, fontsize=30);
```

The bar chart of the variation in prediction accuracy with the year shows that as the years progress, the predicted accuracy for the LDA and Logistic regression models decreases considerably and by 2010, the models are no better than a random classifier. Initially all the models have high accuracy with the LDA and adjusted LDA performing better than logisitic regression. As the number of years from the base year (2000) increases, the accuracy of the logistic model decreases considerably as the class proportions become extremely unequal. The adjusted LDA model far outperforms the LDA and logistic model in terms of accuracy as it is able to adjust for the class proportions in the classification of an observation - the inclusion of the prior allows the model to adjust the decision boundary appropriately to reflect the changing distribution of economic groups 

This analysis suggests that the adjusted LDA model is best suited to model and make predictions on this data set as it uses all the knowledge about the change in the demographics of the population over time.

```python
#--------  plot_decision_boundary
# A function that visualizes the data and the decision boundaries
# Input: 
#      x (predictors)
#      y (labels)
#      poly_flag (a boolean parameter, fits quadratic model if true, otherwise linear)
#      title (title for plot)
#      ax (a set of axes to plot on)
# Returns: 
#      ax (axes with data and decision boundaries)

def plot_decision_boundary(x, y, model, poly_flag, title, ax):
    # Plot data
    ax.scatter(x[y == 1, 0], x[y == 1, 1], c='b', label='class 1')
    ax.scatter(x[y == 0, 0], x[y == 0, 1], c='r', label='class 0')
    
    # Create mesh
    interval = np.arange(0,1,0.01)
    n = np.size(interval)
    x1, x2 = np.meshgrid(interval, interval)
    x1 = x1.reshape(-1, 1)
    x2 = x2.reshape(-1, 1)
    xx = np.concatenate((x1, x2), axis=1)

    # Predict on mesh points
    if(poly_flag):
        quad_features = preprocessing.PolynomialFeatures(degree=2)
        xx = quad_features.fit_transform(xx)
    yy = model.predict(xx)    
    yy = yy.reshape((n, n))

    # Plot decision surface
    x1 = x1.reshape(n, n)
    x2 = x2.reshape(n, n)
    ax.contourf(x1, x2, yy, alpha=0.1)
    
    # Label axes, set title
    ax.set_title(title, fontsize=15)
    ax.set_xlabel('Latitude', fontsize=15)
    ax.set_ylabel('Longitude', fontsize=15)
    
    return ax
```

```python
def evaluate_models(filename):
    
    # load the data
    data = np.loadtxt(filename)
   
    # extract the predictors and the response
    x = data[:, 0:-1]
    y = data[:, -1]
    
    plt.figure(figsize=(15,10))
    fig, ((ax1, ax2, ax3, ax4, ax5)) = plt.subplots(1,5, figsize=(20,5))
    
    # Logisitc regression
    log_model = LogReg()
    log_model.fit(x,y)  
    log_model_score = log_model.score(x,y)
    plot_decision_boundary(x, y, log_model, False, 'Linear Log Reg, Score:' + str(log_model_score), ax1)
    
    # Logistic regression with quadratic effects
    poly = PolynomialFeatures(degree=2)
    x_poly = poly.fit_transform(x)
    log_model_quad = LogReg()
    log_model_quad.fit(x_poly, y)
    log_model_quad_score = log_model_quad.score(x_poly, y)
    plot_decision_boundary(x, y, log_model_quad, True, 'Quadratic Log Reg, Score:' + str(log_model_quad_score), ax2)
    
    # LDA
    lda_model = LDA()
    lda_model.fit(x,y)
    lda_model_score = lda_model.score(x,y)
    plot_decision_boundary(x, y, lda_model, False, 'LDA, Score:' + str(lda_model_score), ax3)
    
    # QDA 
    qda_model = QDA()
    qda_model.fit(x,y)
    qda_model_score = qda_model.score(x,y)
    plot_decision_boundary(x, y, qda_model, False, 'QDA, Score:' + str(qda_model_score), ax4)

    # KNN 
    knn_model = KNN(n_neighbors=5) # arbitrarily chosen number of neighbours
    knn_model.fit(x,y)
    knn_model_score = knn_model.score(x,y)
    plot_decision_boundary(x, y, knn_model, False, 'KNN, Score:' + str(knn_model_score), ax5)
    
    plt.legend(loc='best', bbox_to_anchor=(2,0.75), fontsize=20)
    plt.tight_layout()
```

### Regularisation 

#### Regularising the class proportions

When the training set is small, the estimated class proportions may be noisy and unreflective of the true class proportions.  As the training set increases in size the proportions should tend towards the true proportions of the classes. However, for small training sets, this sampled proportion value may be noisy and thus the model may be improved by regularising the proportions to be equal i.e. $\frac{1}{K}$ where $K$ is the number of classes. The proportions can be regularised using the following form, where $N_i$ is the number of samples in class i, $N$ is the total number of samples in the training set and $\alpha$ is a regularisation parameter:
$$ \hat{\pi_i} = \frac{N_i + \alpha}{N + K\alpha} $$

By the law of large numbers, if the value of $\alpha$ is large, the class proportions will be approximately equal to $\frac{1}{K}$. Different values of alpha can be tried (with different sizes of training sets) and the LDA model fitted with the resulting priors to see how the decision boundary shifts. 

#### Regularising the covariance matrix

A regularisation parameter could be introduced to weight the covariance matrix between LDA and QDA. Both the LDA and QDA classifiers result from assuming that the observations from each class are drawn from a Gaussian distribution. While LDA assumes all classes have a covariance matrix that is common to all K classes, QDA relaxes this assumption and allows each class to have its own covariance. 

The Linear Discriminant Function for LDA is given by:

$$δk(x) = x^T Σ^{-1}μ_k − \frac{1}{2}μ_k^T Σ^{-1}μ_k + log (π_k)$$

where a point is assigned a class with the largest value of the discriminant function. For QDA the discriminant function takes the form:

$$δk(x) = −\frac{1}{2}x^TΣ_k^{−1}x+x^TΣ_k^{−1}μ_k−\frac{1}{2}μ_k^TΣ_k^{-1}μ_k−\frac{1}{2}log|Σ_k|+log(π_k)$$

There is a bias-variance tradeoff in determining which classifier to use. Regularization reduces the variance associated with the estimate at the expense of potentially increased bias. 

Assuming that there are $K$ classes and $p$ predictors, an LDA model will have $Kp$ parameters compared to a QDA model which estimates a separate covariance matrix for each class and consequently has $Kp(p+1)/2$ parameters. The LDA model is therefore less flexible than a QDA model and so has lower variance (desirable when the number of training observations is low). However if the K classes do not share a common covariance matrix, then LDA can suffer from high bias. On the other hand, a QDA model is much more flexible as it relaxes the assumption of a shared covariance matrix but this comes at the risk of overfitting due to the large number of parameters. 

This suggests that a classifier which is somewhere in between an LDA and QDA model will result in a superior performance. A weighted combination of the covariance matrix from LDA and QDA could be used as a form of regularisation, where $\Sigma_k$ is the class specific covariance and $\Sigma$ is the shared covariance estimated from LDA.  

$$ \hat{\Sigma} = \lambda\Sigma + (1-\lambda)\Sigma_k $$

$\lambda$ is a tuning parameter which can range from 0 to 1 and weights more or less of the QDA vs LDA model. When $\lambda = 0$, the regularised covariance is the same as from QDA, and when $\lambda = 1$, the regularised covariance is the same as from LDA. A value between 0 and 1 represents a combination of the two models.

This method of regularisation is used by the sklearn QDA package. This effect of regularisation can be visualised by plotting the decision boundaries for different values of the regularisation parameter. As $\lambda$ varies from 0 to 1, the shape of the boundary changes from quadratic to linear.
